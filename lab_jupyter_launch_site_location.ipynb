```python
# TASK 3: Calculate the distances between a launch site to its proximities


from math import sin, cos, sqrt, atan2, radians
import folium
from folium.plugins import MarkerCluster

# Function to calculate distance between two coordinates (in km)
def calculate_distance(lat1, lon1, lat2, lon2):
    # Earth radius in kilometers
    R = 6373.0
    
    # Convert to radians
    lat1_r = radians(lat1)
    lon1_r = radians(lon1)
    lat2_r = radians(lat2)
    lon2_r = radians(lon2)
    
    # Differences
    dlon = lon2_r - lon1_r
    dlat = lat2_r - lat1_r
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1_r) * cos(lat2_r) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c

# Get unique launch sites
unique_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()
print("=" * 60)
print("LAUNCH SITES AND THEIR COORDINATES")
print("=" * 60)
for idx, row in unique_sites.iterrows():
    print(f"{row['Launch Site']}: Lat={row['Lat']:.5f}, Lon={row['Long']:.5f}")

# Calculate distances between launch sites
print("\n" + "=" * 60)
print("DISTANCES BETWEEN LAUNCH SITES")
print("=" * 60)

for i in range(len(unique_sites)):
    for j in range(i+1, len(unique_sites)):
        site1 = unique_sites.iloc[i]
        site2 = unique_sites.iloc[j]
        dist = calculate_distance(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
        print(f"{site1['Launch Site']} → {site2['Launch Site']}: {dist:.2f} km")

# Find nearest city/coast for each launch site
print("\n" + "=" * 60)
print("DISTANCE TO NEAREST COASTLINE (APPROXIMATE)")
print("=" * 60)

# Approximate coastline coordinates for Florida and California
coastline_points = {
    'Florida East Coast': [28.5, -80.5],
    'Florida Coast (Canaveral)': [28.5, -80.55],
    'California Coast': [34.6, -120.6]
}

for idx, row in unique_sites.iterrows():
    site_name = row['Launch Site']
    site_lat = row['Lat']
    site_lon = row['Long']
    
    # For Florida sites
    if 'CCAFS' in site_name or 'KSC' in site_name:
        coast_coord = [28.5, -80.55]
        dist = calculate_distance(site_lat, site_lon, coast_coord[0], coast_coord[1])
        print(f"{site_name}: ~{dist:.2f} km from Florida coast")
    
    # For California site
    elif 'VAFB' in site_name:
        coast_coord = [34.6, -120.6]
        dist = calculate_distance(site_lat, site_lon, coast_coord[0], coast_coord[1])
        print(f"{site_name}: ~{dist:.2f} km from California coast")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site

import folium
from folium.features import DivIcon
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1_r = radians(lat1)
    lon1_r = radians(lon1)
    lat2_r = radians(lat2)
    lon2_r = radians(lon2)
    dlon = lon2_r - lon1_r
    dlat = lat2_r - lat1_r
    a = sin(dlat / 2)**2 + cos(lat1_r) * cos(lat2_r) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Launch site coordinates (CCAFS LC-40)
launch_site_lat = 28.56230
launch_site_lon = -80.57736

# Points of interest (update with actual coordinates from your map)
points_of_interest = {
    "Coastline": [28.56230, -80.56900],
    "Highway (SR-401)": [28.56500, -80.58000],
    "Railway": [28.57000, -80.58500],
    "City (Cape Canaveral)": [28.38800, -80.60300],
    "Airport": [28.46500, -80.56700]
}

# Create map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="LAUNCH SITE: CCAFS LC-40",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# Add markers for each point of interest with distance line
for name, coord in points_of_interest.items():
    # Calculate distance
    distance = calculate_distance(launch_site_lat, launch_site_lon, coord[0], coord[1])
    
    # Add marker with distance label
    folium.Marker(
        coord,
        popup=f"{name}<br>Distance: {distance:.2f} km",
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, -15),
            html=f'<div style="font-size: 11px; color:blue; background:white; padding:2px; border-radius:5px;"><b>{name}<br>{distance:.2f} km</b></div>'
        )
    ).add_to(site_map)
    
    # Draw line between launch site and point of interest
    folium.PolyLine(
        [[launch_site_lat, launch_site_lon], coord],
        color='purple',
        weight=2,
        opacity=0.6,
        popup=f"Distance: {distance:.2f} km"
    ).add_to(site_map)

# Add Mouse Position
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# Display map
site_map
```
